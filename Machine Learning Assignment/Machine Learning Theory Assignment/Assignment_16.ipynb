
### Q10. In a regression equation, what is multicollinearity ?

**Ans:** 
```python
Multicollinearity occurs when two or more independent variables are highly correlated with one another in a regression model. This means that an independent variable can be predicted from another independent variable in a regression model.

Multicollinearity can be assessed using techniques such as correlation matrices, variance inflation factors (VIF), eigenvalues, or condition indices. If multicollinearity is detected, it can be addressed through techniques like dropping one of the correlated variables, combining them into a composite variable, or using regularization methods like ridge regression or LASSO regression.

Overall, multicollinearity can affect the reliability and interpretability of regression models, so it's important to be aware of it and take appropriate steps to address it if necessary.
```

### Q11. What is heteroskedasticity, and what does it mean ?

**Ans:**  
```python
Heteroskedasticity is a term used in regression analysis to describe a situation where the variability of the error terms (residuals) in a regression model is not constant across all levels of the independent variables. In simpler terms, it means that the spread of the residuals changes as the value of the independent variable(s) changes.
```
**Heteroskedasticity can arise due to various reasons, such as:-**
```python
Presence of outliers or influential observations.
Differences in the scale or variability of the data across different groups or levels of the independent variables.
Omitted variable bias, where important variables that affect the variability of the dependent variable are not included in the model.
```
**To address heteroskedasticity, various techniques can be employed:-**
```python
Transforming the dependent variable or one or more independent variables to stabilize the variance.
Weighted least squares (WLS) estimation, where observations are weighted based on their variance.
Robust standard errors estimation, which adjusts the standard errors of the coefficients to account for heteroskedasticity.
```
